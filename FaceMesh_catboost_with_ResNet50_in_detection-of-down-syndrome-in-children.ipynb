# Детекция Синдрома Дауна у детей

Синдром Дауна — генетическое заболевание, возникающее в результате наличия дополнительной копии 21-й хромосомы. В этом случае сумма хромосом, которая обычно равна 46, увеличивается до 47. Этот дополнительный генетический материал вызывает физические особенности и особенности развития, связанные с синдромом Дауна. Синдром Дауна встречается в среднем у каждых 800 новорожденных. Во всем мире насчитывается около 6 миллионов человек с синдромом Дауна.

Физические характеристики людей с синдромом Дауна могут включать плоский профиль лица, маленькую голову, короткую шею, миндалевидные глаза и высунутый язык. Люди с синдромом Дауна могут быть невысокими и иметь избыточный вес. Возможна степень умственной отсталости, которая может варьироваться от легкой до умеренной.

Характерные черты, обычно сопутствующие синдрому Дауна

Синдром Дауна также может вызывать определенные проблемы со здоровьем, такие как болезни сердца, потеря слуха, проблемы со зрением и проблемы с щитовидной железой.

С этими трудностями люди с синдромом Дауна могут зарабатывать на жизнь. Они могут ходить в школу, работать на работе и строить отношения. При наличии соответствующего образования, поддержки и доступности люди с синдромом Дауна могут реализовать свой потенциал.

Поскольку синдром Дауна не является таким уж редким заболеванием, повышение уровня выявления синдрома во многих отношениях полезно для общества. По этой причине проводились исследования по выявлению синдрома Дауна, хотя и не многие.

```python
def img_prep(path:str, save_to_path: bool=False):
    try:
        image = cv2.imread(path, cv2.COLOR_RGB2GRAY)
    except:
        print('Try another foto')

    def del_background(image):
        # Инициализация FaceMesh
        mp_face_mesh = mp.solutions.face_mesh
        face_mesh = mp_face_mesh.FaceMesh()

        # Обнаружение лица с помощью FaceMesh
        result = face_mesh.process(image)

        if result.multi_face_landmarks:
            for face_landmarks in result.multi_face_landmarks:
                # Получение координат ключевых точек лица
                landmarks = face_landmarks.landmark
                image_height, image_width, _ = image.shape

                # Определение границ области лица
                x_min, y_min = image_width, image_height
                x_max, y_max = 0, 0

                # Построение области вокруг лица
                face_contour = []
                for landmark in landmarks:
                    x, y = int(landmark.x * image_width), int(landmark.y * image_height)
                    face_contour.append((x, y))
                    x_min = min(x_min, x)
                    y_min = min(y_min, y)
                    x_max = max(x_max, x)
                    y_max = max(y_max, y)

                # Обход координат по часовой стрелке
                convex_hull = cv2.convexHull(np.array(face_contour, dtype=np.int32), clockwise=True)

                # Создание маски для лица
                mask = np.zeros((image_height, image_width), dtype=np.uint8)
                cv2.fillPoly(mask, [convex_hull], 255)

                # Применение маски к изображению
                result_image = cv2.bitwise_and(image, image, mask=mask)

                # Обрезка изображения вокруг лица
                face_roi = result_image[y_min:y_max, x_min:x_max]

                # Сохранение результата
                #cv2.imwrite('result_image.jpg', face_roi)

#                 plt.imshow(face_roi)
                return(face_roi)
        else:
            print("No face detected in the image.")
            sys.exit(0)
        # Очистка ресурсов
        face_mesh.close()


    def rotate_img(image):
        mp_drawing = mp.solutions.drawing_utils
        mp_face_mesh = mp.solutions.face_mesh
        # Обрабатываем поворот
        image  = del_background(image)

        with mp_face_mesh.FaceMesh(
            static_image_mode=True,
            max_num_faces=1,
            min_detection_confidence=0.5) as face_mesh:

            # Обработка изображения и поиск ключевых точек на лице
            results = face_mesh.process(cv2.cvtColor(image, cv2.COLOR_BGR2RGB))

            # Если найдено лицо
            if results.multi_face_landmarks:
                # Получение координат ключевых точек глаз
                left_eye_x = results.multi_face_landmarks[0].landmark[159].x * image.shape[1]
                left_eye_y = results.multi_face_landmarks[0].landmark[159].y * image.shape[0]
                right_eye_x = results.multi_face_landmarks[0].landmark[386].x * image.shape[1]
                right_eye_y = results.multi_face_landmarks[0].landmark[386].y * image.shape[0]

                # Вычисление угла поворота лица
                angle = 1 * (180/3.14) * np.arctan((left_eye_y - right_eye_y) / (left_eye_x - right_eye_x))

                # Масштабирование и поворот изображения
                center = (left_eye_x + right_eye_x) // 2, (left_eye_y + right_eye_y) // 2
                scale = 1
                M = cv2.getRotationMatrix2D(center, angle, scale)
                image_aligned = cv2.warpAffine(image, M, (image.shape[1], image.shape[0]))

                return (image_aligned)

            else:
                print('No face detected in the image.')
                sys.exit(0)

        mp_face_mesh.close()
        mp_drawing.close()


    # Сохраняем если нужно
    if save_to_path == True:
        cv2.imwrite(path.split('/')[-1], rotate_img(image))
    else:
        pass

    return rotate_img(image)
```

```python
def evq_dist(point_a:list, point_b:list)->float:
    '''
    Calc Euclidean distance
    
    point_a == [x,y,z];
    point_b == [x,y,z];
    
    return:
    Euclidean distance == float
    '''

    return np.linalg.norm(np.array(point_a)-np.array(point_b))
```

```python
def extr_coord(
    detection_result, 
    part:str,
    num:int=0)->list:
    
    '''
    Extract coordinates, if detection failed return [np.nan]
    
    detection_result == vision.FaceLandmarker result
    part == face landmark name
    num == landmark cord number
    
    result:
    list == [x,y,z]
    '''

    face_landmarks_list = detection_result.face_landmarks
    landmark_points = [FACE_LANDMARKS_NAMES[part][num]]
    if len(detection_result.face_landmarks)>0:
        face_landmarks = [detection_result.face_landmarks[0][i] for i in landmark_points]
        x = face_landmarks[0].x
        y = face_landmarks[0].y
        z = face_landmarks[0].z
        return [x,y,z]
    else:
        return [np.nan,np.nan,np.nan]
```

```python
def extract_relative_dist(detection_result)->pd.Series:
    '''
    Calc relative distance between face key point
    
    detection_result: mediapipe.tasks.python.vision.face_landmarker.FaceLandmarkerResult
    
    return:
    pd.Series([
        iris_diameter, eye_distance, eye_nose_distance,
        nose_mouth_distance, mouth_eye_distance, nose_tip_width,
        eye_brow_distance, face_length, face_width
    ])
    '''
    result = {}

    if extr_coord(detection_result, 'leftEyeIris', 1) != np.nan:
        # iris diameter mean 
        iris_diameter_left = evq_dist(extr_coord(detection_result, 'leftEyeIris', 1), 
                                 extr_coord(detection_result, 'leftEyeIris', 3))
        
        iris_diameter_right = evq_dist(extr_coord(detection_result, 'rightEyeIris', 1), 
                                 extr_coord(detection_result, 'rightEyeIris', 3))
        
        result['iris_diameter_mean'] = np.mean([iris_diameter_left,iris_diameter_right])
    
        iris_diameter = result['iris_diameter_mean']
        
        # beetwin iris
        result['eye_distance'] = evq_dist(extr_coord(detection_result, 'leftEyeIris', 0), 
                                 extr_coord(detection_result, 'rightEyeIris', 0))/iris_diameter
        # midway between eyes and nosetip
        result['eye_nose_distance'] = evq_dist(extr_coord(detection_result, 'midwayBetweenEyes', 0), 
                                 extr_coord(detection_result, 'noseTip', 0))/iris_diameter
        # nosetip and upper outer lip
        result['nose_mouth_distance'] = evq_dist(extr_coord(detection_result, 'noseTip', 0), 
                                 extr_coord(detection_result, 'lipsUpperOuter', 5))/iris_diameter
        # midway between eyes and upper outer lip
        result['mouth_eye_distance'] = evq_dist(extr_coord(detection_result, 'midwayBetweenEyes', 0), 
                                 extr_coord(detection_result, 'lipsUpperOuter', 5))/iris_diameter
        # nose right corner and nose left corner
        result['nose_tip_width'] = evq_dist(extr_coord(detection_result, 'noseRightCorner', 0), 
                                 extr_coord(detection_result, 'noseLeftCorner', 0))/iris_diameter
        # left eyebrow upper and right eyebrow upper
        result['eye_brow_distance'] = evq_dist(extr_coord(detection_result, 'leftEyebrowUpper', -1), 
                                 extr_coord(detection_result, 'rightEyebrowUpper', -1))/iris_diameter
        # up and down face points
        result['face_length'] = evq_dist(extr_coord(detection_result, 'silhouette', 9), 
                                 extr_coord(detection_result, 'silhouette', 27))/iris_diameter
        # left and right face points
        result['face_width'] = evq_dist(extr_coord(detection_result, 'silhouette', 18), 
                                 extr_coord(detection_result, 'leftEyeIris', 0))/iris_diameter
        # mouth width
        result['mouth_width'] = evq_dist(extr_coord(detection_result, 'lipsUpperOuter', 0), 
                                 extr_coord(detection_result, 'lipsUpperOuter', -1))/iris_diameter
        # cheek width
        result['cheek_width'] = evq_dist(extr_coord(detection_result, 'rightCheek', 0), 
                                 extr_coord(detection_result, 'leftCheek', 0))/iris_diameter
        # mouth height
        result['mouth_height'] = evq_dist(extr_coord(detection_result, 'lipsUpperOuter', 5), 
                                 extr_coord(detection_result, 'lipsLowerOuter', 4))/iris_diameter
            
        
        # raw coord
        result['leftEyeIris_1'] = extr_coord(detection_result, 'leftEyeIris', 1)
        result['leftEyeIris_3'] = extr_coord(detection_result, 'leftEyeIris', 3)
        result['leftEyeIris_0'] = extr_coord(detection_result, 'leftEyeIris', 0)
        result['rightEyeIris_1'] = extr_coord(detection_result, 'rightEyeIris', 1)
        result['rightEyeIris_3'] = extr_coord(detection_result, 'rightEyeIris', 3)        
        result['rightEyeIris_0'] = extr_coord(detection_result, 'rightEyeIris', 0)
        
        result['midwayBetweenEyes'] = extr_coord(detection_result, 'midwayBetweenEyes', 0)
        result['noseTip'] = extr_coord(detection_result, 'noseTip', 0)
        
        result['lipsUpperOuter_0'] = extr_coord(detection_result, 'lipsUpperOuter', 0)
        result['lipsUpperOuter_1'] = extr_coord(detection_result, 'lipsUpperOuter', -1)
        result['lipsUpperOuter_4'] = extr_coord(detection_result, 'lipsUpperOuter', 4)
        result['lipsLowerOuter_5'] = extr_coord(detection_result, 'lipsLowerOuter', 5)
        
        result['noseRightCorner'] = extr_coord(detection_result, 'noseRightCorner', 0)
        result['noseLeftCorner'] = extr_coord(detection_result, 'noseLeftCorner', 0)
        
        result['silhouette_0'] = extr_coord(detection_result, 'silhouette', 0)
        result['silhouette_9'] = extr_coord(detection_result, 'silhouette', 9)
        result['silhouette_18'] = extr_coord(detection_result, 'silhouette', 18)
        result['silhouette_27'] = extr_coord(detection_result, 'silhouette', 27)
        
        result['rightCheek'] = extr_coord(detection_result, 'rightCheek', 0)
        result['leftCheek'] = extr_coord(detection_result, 'leftCheek', 0)
        
        result['leftEyebrowUpper'] = extr_coord(detection_result, 'leftEyebrowUpper', -1)
        result['rightEyebrowUpper'] = extr_coord(detection_result, 'rightEyebrowUpper', -1)

        return result
    else:
        print('detection failed')
        result
```

```python
def extra_pic_info(file_name: str) -> pd.Series:
    '''
    Add face key points distance in dataframe
    '''
    # create path to directories
    if file_name.split('_')[0] == 'down':
        path = os.path.join(PIC_PATH, 'downSyndorme', 'downSyndrome', file_name) 
    else:
        path = os.path.join(PIC_PATH, file_name.split('_')[0], file_name.split('_')[0], file_name)
    
    # load pic
    image = mp.Image.create_from_file(path)

    # landmarks extract
    detection_result = detector.detect(image)
    
    # apply func 
    return extract_relative_dist(detection_result)
```

```python
def path_extract(file_name: str) -> pd.Series:
    '''
    Add face key points distance in dataframe
    '''
    # create path to directories
    if file_name.split('_')[0] == 'down':
        path = os.path.join(PIC_PATH, 'downSyndorme', 'downSyndrome', file_name) 
    else:
        path = os.path.join(PIC_PATH, file_name.split('_')[0], file_name.split('_')[0], file_name)
    
    # apply func 
    return path
```

# Tensorflow classification model

```python
def img_preprocessing(path, label):
    
    img = tf.io.read_file(path)
    img = tf.io.decode_jpeg(img, channels = 3)
    img = tf.image.resize(img, size = (250,250)) # mb no resize
    img = img/255.0
    
    return img, label
```

```python
SEED = 15
BATCH_SIZE = 16 #8

tf.keras.utils.set_random_seed(
    SEED
)

# Creating dataset loaders
train_loader = tf.data.Dataset.from_tensor_slices((X_train['path'], y_train))
train_dataset = (train_loader.map(img_preprocessing).batch(BATCH_SIZE).shuffle(X_train['path'].shape[0]).prefetch(BATCH_SIZE))

test_loader = tf.data.Dataset.from_tensor_slices((X_test['path'], y_test))
test_dataset = (test_loader.map(img_preprocessing).batch(BATCH_SIZE).prefetch(BATCH_SIZE))
```

```python
data_augmentation = tf.keras.Sequential(
  [
    layers.RandomFlip("horizontal",
                      input_shape=(250, 250, 3)),
    layers.RandomRotation(0.1), # 
    layers.RandomZoom(0.1),
    layers.RandomContrast(0.1),
    layers.RandomCrop(200, 200, 3),
    layers.Resizing(250, 250)
  ]
)
```

```python
backbone = ResNet50( #ResNet101
    input_shape=(250, 250, 3),
    weights='imagenet',
    include_top=False)
model = Sequential()
model.add(data_augmentation)
model.add(backbone)
model.add(GlobalAveragePooling2D())
model.add(Dense(1, activation='sigmoid'))

model.summary()
```

```python
model.compile(
    optimizer=tf.keras.optimizers.Adam(learning_rate=0.00001),
    loss = 'binary_crossentropy',
    metrics = ['accuracy','AUC','Precision','Recall']
#     metrics = ['binary_accuracy']
)
```

```python
callbacks = [EarlyStopping(monitor = 'val_accuracy', patience = 8, min_delta = 0.001)]
checkpointer = ModelCheckpoint(filepath='bestvalue', verbose=0, save_best_only=True)
reduce_lr = ReduceLROnPlateau(monitor='val_accuracy', factor=0.2, patience=3, min_lr=0.0000001)

callback_list = [checkpointer, callbacks, reduce_lr]

history = model.fit(
    train_dataset,
    validation_data=test_dataset, 
    epochs = 20, 
    verbose =1, 
    callbacks = callback_list)
```

```python
fig, axs = plt.subplots(1,2, figsize = (12,5), dpi = 100)

axs[0].grid(linestyle="dashdot")
axs[0].set_title("Train Set Accuracy")
axs[0].plot(history.history['accuracy'])
axs[0].plot(history.history['val_accuracy'])

axs[0].set_xlabel("epochs")
axs[0].legend(["Accuracy"])


axs[1].grid(linestyle="dashdot")
axs[1].set_title("AUC")
axs[1].plot(history.history['auc'])
axs[1].plot(history.history['val_auc'])
# axs[1].plot(history.history['val_recall'])
axs[1].set_xlabel("epochs")
# axs[1].legend(["AUC", "Precision", "Recall"])
```
